```python
import config2 # to hide TMDB API keys
import requests # to make TMDB API calls
import locale # to format currency as USD
locale.setlocale( locale.LC_ALL, '' )
import json
import pandas as pd

api_key = config2.tmdb_api_key
```

```python
response = requests.get('https://api.themoviedb.org/3/discover/movie?api_key='+ api_key + '&language=en-US&region=US&sort_by=popularity.desc&include_adult=false&include_video=false&page=10')
```

```python
data = response.json()
for result in data['results']:
    columns = list(result.keys())
    break
```

```python
movie_dataframe = pd.DataFrame(columns=columns)
```

```python
movie_dataframe.info()
```

```python
for i in range(200):
    response = requests.get(f'https://api.themoviedb.org/3/discover/movie?api_key={api_key}&language=en-US&region=US&sort_by=popularity.desc&include_adult=false&include_video=false&page={i+1}').json()
    for result in response['results']:
        if result['original_language']=='en':
            movie_dataframe = movie_dataframe.append(result, ignore_index=True)
```

```python
genre_json_file = open('genres.json')

genre_data = json.load(genre_json_file)

genre_list = genre_data['genres']
    
```

```python
def obtaining_genre_from_genre_ids(movie_dataframe,genre_list):
    genre_list_for_dataframe= list()
    for i in range(len(movie_dataframe)):
        genre_ids = movie_dataframe.loc[i,'genre_ids']
        genres = list()
        for i in genre_list:
            if i['id'] in genre_ids:
                genres.append(i['name'])
        genre_list_for_dataframe.append(genres)

    movie_dataframe['genres'] = genre_list_for_dataframe
    
    return movie_dataframe
```

```python
def basedOnNumberCast(top_cast_per_film, num, response):
    for i in range(num):
        top_cast_per_film.append(response['cast'][i]['name'])
    return top_cast_per_film
```

```python
def obtaining_actors_and_actresses(movie_dataframe):
    top_cast_for_dataframe = list()
    movie_ids = movie_dataframe['id']
    
    for movie_id in movie_ids:
        response = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key={api_key}&language=en-US').json()
        top_cast_per_film = list()
        
        if len(response['cast']) >= 10:
           top_cast_per_film = basedOnNumberCast(top_cast_per_film, 10, response)
        else:
            top_cast_per_film = basedOnNumberCast(top_cast_per_film, len(response['cast']), response)

        top_cast_for_dataframe.append(top_cast_per_film)
    
    movie_dataframe['top_cast']=top_cast_for_dataframe
    return movie_dataframe
                
```

```python
movie_dataframe=obtaining_genre_from_genre_ids(movie_dataframe, genre_list)
```

```python
movie_dataframe = obtaining_actors_and_actresses(movie_dataframe)
```
